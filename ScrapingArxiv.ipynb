```python
"""
This code is scrapping Arxiv for research papers by keywords

"""
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
topics = pd.read_csv('PhysicsTopics.csv')
```

```python
df = pd.DataFrame({'links': list(),'topic':list()})
for j in topics['Topics']:
    for i in range(0,10000,200):
        try:
            search = f"https://arxiv.org/search/?searchtype=all&query={j}&abstracts=show&size=200&order=&start={i}"
            response = requests.get(search)
            soup = BeautifulSoup(response.content, 'html.parser')
            x = soup.find_all('a')
            for i in x:
                try:
                    if 'abs' in i['href'] and 'search' not in i['href'] and i['href'] not in df['links']:
                        df.loc[len(df.index)] = [i['href'], j]
                except:
                    continue
        except:
            continue
    df.to_csv('arxivLinksPhysics.csv',index=False)
    print(j)
    
```

```python
Links = pd.read_csv('arxivLinksPhysicsNew.csv')
```

```python
Links['links'][0][:8]+'export.'+Links['links'][0][8:]
```

```python

for i in range(len(Links['links'])):
    try:
        if pd.isna(Links['abstract'][i]):
            link = Links['links'][i][:8]+'export.'+Links['links'][i][8:]
            Research = requests.get(link)
            researchSoup = BeautifulSoup(Research.content, 'html.parser')
            y = researchSoup.find('blockquote', class_="abstract mathjax")
            links = researchSoup.find_all('a')
            if 'Abstract' in y.text:
                Links['abstract'][i] = y.text[11:].strip()
                for j in links:
                    try:
                        if 'doi' in j['href']:
                            Links['DOI'][i] = j['href']
                    except:
                        continue
        else:
            continue
        if i % 100 == 0:
            Links.to_csv('arxivLinksPhysicsNew.csv',index=False)
            print(i)
    except:
        continue

```
